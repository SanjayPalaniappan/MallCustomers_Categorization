```python
# Importing the required python libraries
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# Reading the input data present in csv file and storing it into a variable & printing the same
dataset = pd.read_csv('Mall_Customers.csv')
dataset
```

```python
# Considering only Annual Income and Spending Score columns, hence storing those in separate variable for ease
X = dataset.iloc[:, 3:5].values
X
```

```python
# Identifying value of 'k' using Elbow method
from sklearn.cluster import KMeans
list1 = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters = i, init = 'k-means++', random_state = 42)
    kmeans.fit(X)
    list1.append(kmeans.inertia_)
plt.plot(range(1, 11), list1)
plt.title('The Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```

```python
# Printing list contaning kmeans_inertia values
list1
```

```python
# Building kmeans model using 'k' value identified through Elbow method and predicting model using input data
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters = 5, init = 'k-means++', random_state = 42)
y_kmeans = kmeans.fit_predict(X)
y_kmeans
```

```python
#To add new column, assigning dataset to new variable and printing the same
supervised=dataset
supervised
```

```python
# Adding a new column named 'cluster_group' into dataset
supervised['Cluster_group']=y_kmeans
supervised
```

```python
# Putting the dataset with clusters segmented using kmeans into a csv file
supervised.to_csv("Kmeans_clusters.csv",index=False)
```
